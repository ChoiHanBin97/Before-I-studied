## sort(), sorted() 
- sort() : 원본 변경
- sorted() : 대상 변경 없이, 정렬된 결과 리턴

※위 정렬 함수는 `Timsort` 라는 정렬 알고리즘 사용


```python
import random
myList = [
    random.randrange(1, 1000)
    for i in range(1000)
]

# myList
```

```python
myList.sort()   # myList 원본을 변화시켰다!
```

```python
myList[0:20]
```

```python
random.shuffle(myList)
myList[0:20]
```

```python
sorted(myList)     # 정렬된 결과를 리턴!. myList 원본은 변화 안시킴
```

```python
myList
```

# 정렬 알고리즘의 성능
time complexity 는 다음 3가지로 귀결

1. **O(*n*<sup>2</sup>)**  <-- worst
1. **O(*n* log *n*)**    <-- average
1. **O(*n*)**          <-- impossible?

> 특정 정렬 알고리즘의 성능이 항상 고정되어 있지 않다.   정렬하는 데이터의 분포 나 형태에 따라 성능이 달라질수 있다.     따라서,  각 알고리즘을 정확히 이해하고, 구현할줄 알며, 각각의 특성에 따라 어떠한 데이터 상황에서 정렬에 유리한지 알아야 한다.
