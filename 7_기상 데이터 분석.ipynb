### 기상 데이터 분석하기

```python
input_file = "data.csv"
output_file = "tem10y.csv"
```

```python
# CSV 파일을 한 줄씩 읽어오기
with open(input_file, "rt", encoding='EUC-KR') as fr:
    lines = fr.readlines()
```

```python
# 기존의 데이터를 분리해서 가공하기
lines = ["연,월,일,기온,품질,균질\n"]+lines[5:] # 기존헤더 줄을 제거하고 새로운 헤더를 추가
lines = map(lambda v: v.replace("/", ","), lines)
result = "".join(lines).strip()
print(result)
```

```python
with open(output_file, "wt", encoding="utf-8") as fw:
    fw.write(result)
    print("저장되었습니다.")
```

```python
import pandas as pd

df = pd.read_csv("tem10y.csv", encoding="utf-8")
print(df)
```

### 평균 기온 구하기
- 10년 동안의 데이터를 집계해서 각 연월의 평균 기온을 구합니다. 

```python
for i, row in df.iterrows():
    print(row)
    print(type(row))
    print(i, '번째')
```

```python
# array는 []를 쓰는 대표적인 타입입니다. []는 배열을 선언 및 초기화할 때 배열의 원소에
# 접근할 떄 사용합니다.
# Tuple은 ()를 쓰는 대표적인 타입입니다. ()는 튜플을 선언 및 초기화할 때 사용합니다.
# []는 튜플의 원소에 접근할 때 사용합니다.
# Dictionary는 {}를 쓰는 대표적인 타입입니다. {}는 딕셔너리를 선언 및 초기화할 때 사용합니다.
# {}는 Key에 대응하는 value를 할당하거나 value에 접근할 때 쓰입니다.



md = {}
# DataFrame 자료형의 데이터를 한 줄씩 처리하고 싶을 때 for ~ df.iterrows() 메소드를 사용
for i, row in df.iterrows():
    m, d, v = (int(row['월']), int(row['일']), float(row['기온']))
    key = str(m) + "/" + str(d)
    if not(key in md): md[key] = []
    md[key] += [v]

avs = {}
for key in md:
    v = avs[key] = sum(md[key]) / len(md[key])
    print("{0} : {1}".format(key, v))
```

```python
# 월별 평균 기온 구하기
# 날짜별 평균은 DataFrame 자료형의 데이터를 한 행씩 처리하여 값을 구했습니다.
# Pandas 모듈을 이용하면 DataFrame의 groupby() 메소드를 사용해 간단하게 집계를 할 수 있습니다.

g = df.groupby(['월'])['기온']    # 월의 기온을 가져옴
print(g, 'g를 불렀습니다.', sep = '\n')
gg = g.sum() / g.count()
print(gg)

import matplotlib.pyplot as plt
gg.plot()
plt.savefig('tmp-month-avg.png')
plt.show()

# 출력
'''
월
1      5.996481
2      6.598714
3     10.017009
4     14.481515
5     19.607918
6     22.544848
7     26.422287
8     27.887097
9     24.360303
10    19.155132
11    13.484545
12     8.653079
'''
```

```python
# 기온이 30도 넘는 날 구하기
# 평균 기온이 30도를 넘는 날이 한 해에 몇번 있었는지 확인

# 기온이 30도를 넘는 데이터 확인
hot_bool = (df['기온'] > 30)
# print(hot_bool)
hot = df[hot_bool]
# print(hot)
cnt = hot.groupby(['연'])['연'].count()
print(cnt)

cnt.plot()
plt.savefig('tem-over30.png')
plt.show()



'''
연
2006     2
2007    11
2008     5
2010    21
2011     9
2012     8
2013    16
2014    12
2015     7
2016     1
'''
```
