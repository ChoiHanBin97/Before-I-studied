### 머신러닝 프로세스
- 기획 -> 데이터 수집 -> 데이터 탐색 -> 데이터 전처리 -> 모델링 -> 모델 평가 -> 결과 보고서

#### 데이터 수집 -> 크롤링(스크레이핑)
##### 웹 크롤러(Web Crawler)
- 자동화된 방법으로 웹에서 다양한 정보를 수집하는 소프트웨어를 말합니다.
- HTML 문서를 가져오려면 웹 사이트의 URL로 접속하여 데이터를 가져옵니다. (requests 모듈)
- HTML 문서를 파싱(분석)하려면 BeautifulSoup 모듈을 설치해야합니다.

##### 웹페이지
- HTML : 웹사이트의 틀
- CSS : HTML을 보완하여 디자인적인 영역을 담당
- JavaScript : 동적인 프로그램(사용자에 대해서 반응해주는 부분 등)



```python
# 크롤링 사이트
# https://basicenglishspeaking.com/daily-english-conversation-topics/

# pip install requests
# pip install BeautifulSoup4

import requests 
from bs4 import BeautifulSoup
```

```python
def getSubject():
    # print('getSubject() 함수 호출')
    subject = [] # 파싱된 데이터를 저장할 빈 List를 만듭니다.
    request = requests.get('https://basicenglishspeaking.com/daily-english-conversation-topics/')
    # print(requests) # <Response [200]> : 정상적인 연결, <Response [404]> : 페이지가 없음
    html = request.text # 요청한 페이지 정보 중에서 문자열만 저장합니다.
    # print(html)
    
    # BeautifulSoup(파싱할 HTML 소스, 파싱에 사용될 파서)
    # HTML문서를 파싱 : html.parser / xml문서를 파싱 : xml.parser
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    # print(type(soup))
    divs = soup.findAll('div', {"class":"thrv-columns"})
    # print(divs)
    
    for div in divs:
        links = div.findAll("a")
        # print(links)
        for link in links:
            # print(link.text)
            subject.append(link.text)
    # print(subject)
    return subject

if __name__ == "__main__" :   # 파이썬의 프로그램 시작점을 만듭니다.
    # print("파이썬 프로그램의 시작")
    result = getSubject()
    # print(result)
    print("총", len(result), "개의 대화를 찾았습니다.")
    
    for i in range(len(result)):
        print('{0:2d}, {1:s}' .format(i + 1, result[i]) )
    
```
